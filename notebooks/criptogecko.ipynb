```python
import requests
import dotenv
import os

dotenv.load_dotenv()

API_KEY = os.getenv("COINGECKO_API_KEY")

COIN_IDS = "bitcoin,ethereum,binancecoin,solana,ripple,dogecoin,cardano,avalanche-2,shiba-inu,polkadot"

URL = "https://api.coingecko.com/api/v3/coins/markets"

params = {
    "vs_currency": "usd",
    "ids": COIN_IDS,
    "order": "market_cap_desc",
    "price_change_percentage": "24h",
}

headers = {"x-cg-demo-api-key": API_KEY}

response = requests.get(URL, headers=headers, params=params)
response.json()
```

# Features Description

| Field | Type | Description |
|---|---|---|
| `id` | string | CoinGecko unique identifier (e.g. `bitcoin`) |
| `symbol` | string | Ticker symbol (e.g. `btc`) |
| `name` | string | Full coin name (e.g. `Bitcoin`) |
| `image` | string | URL of the coin logo image |
| `current_price` | float | Current price in USD |
| `market_cap` | float | Market capitalization in USD |
| `market_cap_rank` | integer | Global ranking position by market cap |
| `fully_diluted_valuation` | float | Valuation assuming full token supply in circulation |
| `total_volume` | float | Total trading volume in the last 24h in USD |
| `high_24h` | float | Highest price in the last 24h in USD |
| `low_24h` | float | Lowest price in the last 24h in USD |
| `price_change_24h` | float | Absolute price change in the last 24h in USD |
| `price_change_percentage_24h` | float | Percentage price change in the last 24h |
| `market_cap_change_24h` | float | Absolute market cap change in the last 24h in USD |
| `market_cap_change_percentage_24h` | float | Percentage market cap change in the last 24h |
| `circulating_supply` | float | Number of coins currently in circulation |
| `total_supply` | float | Total number of coins in existence |
| `max_supply` | float | Maximum number of coins that can ever exist |
| `ath` | float | All Time High price in USD |
| `ath_change_percentage` | float | Percentage distance from the ATH |
| `ath_date` | timestamp | Date when the ATH was reached |
| `atl` | float | All Time Low price in USD |
| `atl_change_percentage` | float | Percentage distance from the ATL |
| `atl_date` | timestamp | Date when the ATL was reached |
| `roi` | float | Return on investment (frequently null) |
| `last_updated` | timestamp | Timestamp of the last data update |
| `price_change_percentage_24h_in_currency` | float | 24h percentage price change expressed in the base currency (same as `price_change_percentage_24h` for USD) |

```python
import pandas as pd

data = response.json()
df = pd.DataFrame(data)
df["ath_date"] = pd.to_datetime(df["ath_date"])
df["atl_date"] = pd.to_datetime(df["atl_date"])
df["last_updated"] = pd.to_datetime(df["last_updated"])
df
```
