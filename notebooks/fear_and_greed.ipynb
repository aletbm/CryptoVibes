```python
import requests
import dotenv

dotenv.load_dotenv()

URL = "https://api.alternative.me/fng/?limit=0"

response = requests.get(URL)
response.json()
```

# Features Decription

| Field | Type | Description |
|---|---|---|
| `value` | string | Fear and Greed score from 0 (Extreme Fear) to 100 (Extreme Greed), returned as string |
| `value_classification` | string | Human-readable sentiment label: Extreme Fear, Fear, Neutral, Greed, Extreme Greed |
| `timestamp` | string | Unix timestamp representing the date this value corresponds to |
| `time_until_update` | string | Seconds remaining until the next index update (only returned when limit=1) |

```python
import pandas as pd

data = response.json()
df = pd.DataFrame(data["data"])
df["timestamp"] = pd.to_datetime(df["timestamp"].astype(int), unit="s", utc=True)
df = df.drop(columns=["time_until_update"], errors="ignore")
df
```
